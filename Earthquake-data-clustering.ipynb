DATA TRAINING

```python
x_train = gempa[['depth','mag']].values
print(x_train)
```

FEATURE SCALLING

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
x_train = scaler.fit_transform(x_train)

x_train
```

CLUSTERING

```python
from sklearn.cluster import KMeans

xcluster = []
for i in range(1,11):
  kmeans_cek = KMeans(n_clusters = i, init = 'k-means++', random_state = 42)
  kmeans_cek.fit(selected_columns)
  xcluster.append(kmeans_cek.inertia_)
plt.plot(range(1,11), xcluster)
```

karena metode elbow menunjukan elbow mendekati antara angka 2 - 4, maka k = 3

```python
from sklearn.preprocessing import StandardScaler

kmean = KMeans(n_clusters=3)
kmean
```

```python
y_cluster = kmean.fit_predict(x_train)
y_cluster
```

```python
len(np.unique(y_cluster))
```

```python
gempa['cluster'] = y_cluster
gempa.head(5)
```

```python
# 3 centroid
kmean.cluster_centers_
```

```python
plt.figure(figsize=(10,5))
plt.scatter(x_train[:,0],x_train[:,1],c=kmean.labels_)
plt.scatter(kmean.cluster_centers_[0,0],kmean.cluster_centers_[0,1],cmap='rainbow', marker = '*', s = 150, label = 'Centroid 1')
plt.scatter(kmean.cluster_centers_[1,0],kmean.cluster_centers_[1,1],cmap='rainbow', marker = '*', s = 150, label = 'Centroid 2')
plt.scatter(kmean.cluster_centers_[2,0],kmean.cluster_centers_[2,1],cmap='rainbow', marker = '*', s = 150, label = 'Centroid 3')

plt.xlabel('depth')
plt.ylabel('mag')
plt.title('Clustering data gempa dengan 3 cluster')
plt.legend()
```
